**Pasos de la búsqueda binaria:**

Calcular el punto medio de la lista.

Si el valor buscado es menor que el valor en el punto medio de la lista, establecer un nuevo límite derecho.

Si el valor buscado es mayor que el valor en el punto medio de la lista, establecer un nuevo límite izquierdo.

Si el valor buscado es igual al valor en el punto medio de la lista, devolver el punto medio (el índice).

Repetir los pasos anteriores hasta que se encuentre el valor o el límite izquierdo sea igual o mayor que el límite derecho.

Es importante entender que un algoritmo que debe acceder a todos los elementos de sus datos de entrada no puede tener un tiempo logarítmico, ya que el tiempo necesario para leer una entrada de tamaño n es del orden de n.

```python
# 2. Complejidad Logarítmica: O(log n)
def get_first(data):
    return data[0]

if __name__ == '__main__':
    data = [1, 2, 9, 8, 3, 4, 7, 6, 5]
    print(get_first(data))

def logaritmica_algo(n):
    pasos = 0
    i = 1
    while i < n:
        i *= 2
        pasos += 1
    return pasos

for index in range(0, len(data), 3):
    print(data[index])


def binary_search(data, value):
    n = len(data)
    left = 0
    right = n - 1
    while left <= right:
        middle = (left + right) // 2
        if value < data[middle]:
            right = middle - 1
        elif value > data[middle]:
            left = middle + 1
        else:
            return middle
    raise ValueError('Value is not in the list')

if __name__ == '__main__':
    data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    print(binary_search(data, 8))

# Definir el rango de tamaños de entrada
tamaños = [2**i for i in range(12)]
print (tamaños)
# Medir el número de pasos para cada algoritmo y tamaño de entrada
pasos_logaritmica = [medir_pasos(logaritmica_algo, n) for n in tamaños]

# Graficar los resultados
plt.figure(figsize=(10, 8))
plt.plot(tamaños, pasos_logaritmica, label='O(log n) - Logarítmica')

#plt.yscale('log')  # Escala logarítmica para visualizar mejor las diferencias
#plt.xscale('log')
plt.xlabel('Tamaño de entrada (n)')
plt.ylabel('Número de pasos')
plt.title('Comparación de Complejidades Algorítmicas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 4. Complejidad Log-lineal: O(n log n)
def loglineal_algo(n):
    pasos = 0
    i = 1
    while i < n:
        for j in range(n):
            pasos += 1
        i *= 2
    return pasos


def merge_sort(data):
    if len(data) <= 1:
        return

    mid = len(data) // 2
    left_data = data[:mid]
    right_data = data[mid:]

    merge_sort(left_data)
    merge_sort(right_data)

    left_index = 0
    right_index = 0
    data_index = 0

    while left_index < len(left_data) and right_index < len(right_data):
        if left_data[left_index] < right_data[right_index]:
            data[data_index] = left_data[left_index]
            left_index += 1
        else:
            data[data_index] = right_data[right_index]
            right_index += 1
        data_index += 1

    if left_index < len(left_data):
        del data[data_index:]
        data += left_data[left_index:]
    elif right_index < len(right_data):
        del data[data_index:]
        data += right_data[right_index:]

if __name__ == '__main__':
    data = [9, 1, 7, 6, 2, 8, 5, 3, 4, 0]
    merge_sort(data)
    print(data)
# Definir el rango de tamaños de entrada
tamaños = [2**i for i in range(12)]
print (tamaños)
# Medir el número de pasos para cada algoritmo y tamaño de entrada
pasos_loglineal = [medir_pasos(loglineal_algo, n) for n in tamaños]
# Graficar los resultados
plt.figure(figsize=(10, 8))
plt.plot(tamaños, pasos_loglineal, label='O(n log n) - Log-lineal')

#plt.yscale('log')  # Escala logarítmica para visualizar mejor las diferencias
#plt.xscale('log')
plt.xlabel('Tamaño de entrada (n)')
plt.ylabel('Número de pasos')
plt.title('Comparación de Complejidades Algorítmicas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 7. Complejidad Exponencial: O(2^n)
def exponencial_algo(n):
    if n == 0:
        return 1
    else:
        return exponencial_algo(n - 1) + exponencial_algo(n - 1)


def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

# Para evitar tiempos de ejecución muy largos, limitamos la medición de pasos exponenciales
tamaños_exponencial = [2, 4, 6, 8, 10]
pasos_exponencial = [medir_pasos(exponencial_algo, n) for n in tamaños_exponencial]
# Graficar los resultados
plt.figure(figsize=(10, 8))
plt.plot(tamaños_exponencial, pasos_exponencial, label='O(2^n) - Exponencial')

plt.yscale('log')  # Escala logarítmica para visualizar mejor las diferencias
plt.xscale('log')
plt.xlabel('Tamaño de entrada (n)')
plt.ylabel('Número de pasos')
plt.title('Comparación de Complejidades Algorítmicas')
plt.legend()
plt.grid(True)
plt.show()
```
